### This Data Analysis notebook aims to uncover patterns, trends, and insights on public transportation data using Python.

```python
## Import all necessary libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Read in data
data = pd.read_csv('/drive/data/Monthly_Transportation_Statistics.csv')
```

# Visualizing Our Data in different ways

## Comparing Airline Traffic to Transit Ridership throughout the years

```python
# Line plot of U.S. airline traffic - total (seasonally adjusted) vs. transit ridership - fixed route bus (adjusted)
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['U.S. Airline Traffic - Total - Seasonally Adjusted'], label='Airline Traffic')
plt.plot(data['Date'], data['Transit Ridership - Fixed Route Bus - Adjusted'], label='Fixed Route Bus Ridership')
plt.title('U.S. Airline Traffic vs. Transit Ridership')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

## Lets investigate the sharp decrease in Air Traffic and Transit Ridership 

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['U.S. Airline Traffic - Total - Seasonally Adjusted'], label='Airline Traffic')
plt.plot(data['Date'], data['Transit Ridership - Fixed Route Bus - Adjusted'], label='Fixed Route Bus Ridership')

# Highlight the data between the two dates
plt.axvspan('01/01/2020 12:00:00 AM', '05/01/2021 12:00:00 AM', color='grey', alpha=0.5)

plt.title('U.S. Airline Traffic vs. Transit Ridership')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

## The highlighted decline in the graph first occurs March 2020. This is when the Covid Lockdown happened. The lockdown caused a massive drop in Airline Traffic and Transit Ridership

# Comparing Mileages on different highway systems

```python
# Line plot of U.S. airline traffic - total (seasonally adjusted) vs. transit ridership - fixed route bus (adjusted)
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Highway Vehicle Miles Traveled - All Systems'], label='All Systems')
plt.plot(data['Date'], data['Highway Vehicle Miles Traveled - Total Rural'], label='Total Rural')
plt.plot(data['Date'], data['Highway Vehicle Miles Traveled - Other Rural'], label='Other Rural')
plt.plot(data['Date'], data['Highway Vehicle Miles Traveled - Rural Other Arterial'], label='Rural Other Arterial')
plt.plot(data['Date'], data['Highway Vehicle Miles Traveled - Rural Interstate'], label='Rural Interstate')
plt.title('Highway Vehicle Miles Traveled')
plt.xlabel('Date')
plt.ylabel('Miles')
plt.legend()
plt.show()
```

# Comparing incoming persons and truck crossings for US-Mexico and US-Canada border

```python
# Line plot of U.S. airline traffic - total (seasonally adjusted) vs. transit ridership - fixed route bus (adjusted)
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['U.S.-Canada Incoming Person Crossings'], label='U.S.-Canada Person Crossing')
plt.plot(data['Date'], data['U.S.-Canada Incoming Truck Crossings'], label='U.S.-Canada Truck Crossing')
plt.plot(data['Date'], data['U.S.-Mexico Incoming Person Crossings'], label='U.S.-Mexico Incoming Person Crossing')
plt.plot(data['Date'], data['U.S.-Mexico Incoming Truck Crossings'], label='U.S.-Mexico Incoming Truck Crossing')

plt.title('Border Traffic from People and Truck Crossings')
plt.xlabel('Date')
plt.ylabel('Traffic')
plt.legend()
plt.show()
```

# Graphing sales of different vehicle types throughout the years

```python
# Line plot of U.S. airline traffic - total (seasonally adjusted) vs. transit ridership - fixed route bus (adjusted)
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Light truck sales'], label='Light Truck Sales')
plt.plot(data['Date'], data['Auto sales'], label='Auto Sales')
plt.plot(data['Date'], data['Heavy truck sales'], label='Heavy Truck Sales')

plt.title('Light Truck, Auto, and Heavy Truck Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Traffic')
plt.legend()
plt.show()
```
